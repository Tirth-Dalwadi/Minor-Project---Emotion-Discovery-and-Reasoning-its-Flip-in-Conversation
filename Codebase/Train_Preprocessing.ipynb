```python
import re
import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

nltk.download('punkt')
nltk.download('stopwords')
nltk.download('wordnet')

def preprocess_text(text):
    # Lowercasing: Convert text to lowercase
    text = text.lower()

    # Tokenization: Split the text into words
    words = nltk.word_tokenize(text)

    # Remove punctuation and non-alphanumeric characters
    words = [word for word in words if re.match(r'^[a-z]+$', word)]

    # Remove English stopwords
    stop_words = set(stopwords.words('english'))
    words = [word for word in words if word not in stop_words]

    # Lemmatization: Reduce words to their base form
    lemmatizer = WordNetLemmatizer()
    words = [lemmatizer.lemmatize(word) for word in words]

    # Join the words back into a single string
    preprocessed_text = ' '.join(words)

    return preprocessed_text
```

```python
df['cleaned_utterances'] = df['utterances'].apply(preprocess_text)
df
```

```python
from indic_transliteration import sanscript

# Function to transliterate English text to Hindi
def transliterate_to_hindi(text):
    # Specify the source and target scripts
    source_script = sanscript.ITRANS
    target_script = sanscript.DEVANAGARI

    # Transliterate the text
    transliterated_text = sanscript.transliterate(text, source_script, target_script)
    return transliterated_text
```

```python
df['hindi_utterances'] = df['utterances'].apply(transliterate_to_hindi)
```

```python
from mtranslate import translate

def translate_to_english(text):
  english_translation = translate(text, 'en', 'hi')

  return english_translation
```
