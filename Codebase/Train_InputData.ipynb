```python
import pandas as pd
test1 = pd.read_json(file_path)
df = pd.DataFrame(test1)
df
```

```python
len(df['speakers'][1])
```

```python
newDF = pd.DataFrame(columns = ['episode', 'speakers', 'utterances', 'emotions'])

for col in ['speakers', 'utterances', 'emotions']:
    i=0
    for row in range(len(df)):
        tempList = df.iloc[row][col]
        episodeName = df.iloc[row]['episode']
        for e in tempList:
            newDF.at[i, col] = e
            newDF.at[i, 'episode'] = episodeName
            i = i + 1
```
